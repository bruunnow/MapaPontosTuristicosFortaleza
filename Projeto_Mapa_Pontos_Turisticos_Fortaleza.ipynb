```python
## Começamos importando as bibliotecas necessarias para o projeto.

import folium
from folium import plugins
```

```python
## Esse codigo tem o intuito de apresentar os principais pontos turisticos da cidade de fortaleza através de um mapa interativo.

## Primeiro buscamos quais pontos irão se apresentados e guardamos eles.

pontos_turisticos_fortaleza_nomes = ['Dragao do Mar',
                                     'Mercado Central',
                                     'Catedral',
                                     'Theatro Jose de Alencar',
                                     'Passeio Público',
                                     'Feirinha de Beira mar',
                                     'Estação das Artes',
                                     'Espigão do Náutico',
                                     'Fortaleza de Nossa Senhora da Assunção',
                                     'Museu do Ceará',
                                     'Mercado Cultural dos Pinhões',
                                     'Mercado dos Peixes de Fortaleza',
                                     'Praia do Futuro']

## Depois temos que inserir as coordenadas geograficas desses lugares na ordem em que foram inseridos os nomes.

pontos_turisticos_coordenadas = [[-3.7212789465272467, -38.52045556603383],
                                 [-3.723306856579942, -38.52406228870501],
                                 [-3.7246128851204707, -38.52478036233629],
                                 [-3.7274472550519784, -38.53169406233644],
                                 [-3.7226328537096514, -38.526445120007494],
                                 [-3.72521807908773, -38.49575040466528],
                                 [-3.722269809413445, -38.52996461815865],
                                 [-3.724877127097351, -38.49477273534951],
                                 [-3.7225301660749035, -38.52474692000737],
                                 [-3.726149011623577, -38.52656148932311],
                                 [-3.72681427945517, -38.51533358932315],
                                 [-3.7213139284969348, -38.479776831652195],
                                 [-3.7489116084286884, -38.44706081364993]]

## Vamos criar o titulo do mapa

titulo = 'Mapa de Pontos Turísticos de Fortaleza'
titulo_html = """<h1 align = center>{}</h1>""".format(titulo)


## Depois de inserir os nomes dos pontos, suas coordenadas o titulo , iremos criar o mapa.

mapa_pontos_turisticos_fortaleza = folium.Map (location= [-3.7324, -38.4922],
                                               zoom_start= 14,
                                               control_scale=True)


## Adicionamos o titulo em html

mapa_pontos_turisticos_fortaleza.get_root ().html.add_child (folium.Element(titulo_html))


## Com mapa já criado, vamos mudar o tipo de apresentação do mapa.
## Peguei o modelo de apresentação de mapa do site leaflet-extras.github.io/leaflet-providers/preview

folium.TileLayer (tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
                  attr = 'Tiles &copy; Esri &mdash; Source: Esri, i-cubed, USDA, USGS, AEX, GeoEye, Getmapping, Aerogrid, IGN, IGP, UPR-EGP, and the GIS User Community',
                  name = 'Esri.WorldImagery',).add_to (mapa_pontos_turisticos_fortaleza)


## Agora precisamos inserir os marcadores indicando onde estão os pontos turísticos.
## Extraí o modelo de icones do site getbootstrap.com/docs/3.3/components

for i in range(len(pontos_turisticos_coordenadas)):
    coordenada = pontos_turisticos_coordenadas [i]
    maravilha = pontos_turisticos_fortaleza_nomes [i]
    folium.Marker(coordenada, popup= maravilha, icon = folium.Icon( icon= 'glyphicon glyphicon-camera',
                                                                          color= 'blue',
                                                                          icon_color= 'white',
                                                                          prefix= 'glyphicon')).add_to (mapa_pontos_turisticos_fortaleza)


## Agora vamos ajustar a centralização do mapa escolhendo um ponto e adicionando
## na locação do mapa.

mapa_pontos_turisticos_fortaleza.add_child (folium.LatLngPopup())


## Agora iremos salvar o mapa para ser visualizado fora do notebook.

mapa_pontos_turisticos_fortaleza.save ('pontos_turisticos_fortaleza.html')

## Mapa pronto para ser exibido.

mapa_pontos_turisticos_fortaleza
```
